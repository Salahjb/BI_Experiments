```python
import matplotlib.pyplot as plt

# initializing the data for the line plot
x = [10, 20, 30, 40]
y = [20, 25, 35, 55]

# plotting the data (Line plot)
plt.plot(x, y)
plt.title("Line Plot Example")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.show()
```
